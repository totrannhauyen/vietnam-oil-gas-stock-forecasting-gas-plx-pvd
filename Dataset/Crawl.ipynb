```python
import yfinance as yf
import pandas as pd

stocks = {
    "GAS": "GAS.VN",
    "PLX": "PLX.VN",
    "PVD": "PVD.VN"
}

all_data = []

for name, ticker in stocks.items():
    print(f"Đang lấy {name}...")

    df = yf.download(ticker, start="2020-01-01", end="2026-04-23")

    if df.empty:
        print(f"{name} không có dữ liệu")
        continue

    # Bỏ level ticker
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.droplevel(1)

    df.reset_index(inplace=True)

    df.to_csv(f"{name}.csv", index=False)
    print(f"Crawl dữ liệu {name} thành công")

    all_data.append(df)

print("Hoàn thành!")
```
